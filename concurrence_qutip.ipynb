```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import*
```

```python
# We define operators for one qubit

sx = sigmax()
sy = sigmay()
sz = sigmaz()
sp = sigmap()
sm = sigmam()
I = qeye(2)
```

```python
#basis for one qubit
zero = basis(2,0)
one = basis(2,1)
```

```python
#observables to be measured in the experiment
s_zz = tensor(sz,sz)
s_z0 = tensor(sz,I)
s_0z = tensor(I,sz)
```

```python
#basis for the bipartite system
ket_00 = tensor(zero,zero)
ket_11 = tensor(one,one)
```

```python
#list for theta
n = 40
s = [i * 2 * np.pi / n for i in range(n)]
```

```python
#list of states
states = []
for theta in s:
    states.append(np.cos(theta)*ket_00+np.sin(theta)*ket_11)
```

```python
#expectation values
#<sigma_zsigma_z>
sz_sz = expect(s_zz,states)

#<sigma_0sigma_z>
s0_sz = expect(s_0z,states)

#<sigma_zsigma_0>
sz_s0 = expect(s_z0,states)
    
```

```python
#list for C^2
c_sq = []
for i in range(40):
    c_sq.append(0.5*(1+sz_sz[i]**2-s0_sz[i]**2-sz_s0[i]**2))
```

```python
x = np.linspace(0,2*np.pi,100)
```

```python
def f(x):
    return np.sin(2*x)**2;
```

```python
fig, ax = plt.subplots()
ax.plot(x, f(x),color='pink',linestyle='solid', linewidth=3.5 ,label="$C^2_{an}$")
errorbar(s,c_sq,fmt='m.',capsize=5,label=r"$C^2_{num}$",marker='.',markersize=15)
ax.set_title("Concurrence for state $|\psi>=Cos(\Theta)|00>+Sin(\Theta)|11>$", size=14)
#xlabel(r"$\phi$",fontsize=20);
#ylabel(r"$P$",fontsize=20);
xticks(fontsize=10); #s,rotation='horizontal',
yticks(fontsize=10);
legend(fontsize=10);
plt.xlabel("$\Theta$",size=14)
plt.ylabel("$C^2$",size=14)
#ax.set_xlim(-0.05, 6)
#ax.set_ylim(0, 1)
grid("on")
plt.legend()

fig1 = plt.gcf()
plt.show()
plt.draw()
#fig1.savefig('fig9.png', dpi=1000)
```
